*c Detect duplicates and remove them *

According to the dataset, a duplicate row would be the one with the same name, address and location. Hence, duplicates are detected and dropped using drop_duplicates on the name, address and location column. A new dataframe is formed to retain the main dataframe for use if needed.

```python
df_unique=(df.drop_duplicates(subset=["name","address","location"],keep="first")).copy()
df_unique = df_unique[['location','rate','rest_type','cuisines','approx_cost(for two people)']]
```

*d Neighborhood with highest average rating*

The neighborhood with highest average rating is Rajarajeshwari Nagar with a rating of 3.85.
The major characteristics of this rating are that is famously a Casual Dining restaurant, mostly famous for Chinese, Italian, Mexican and Thai food. Also, this area does have a lot of delivery options. The normal cost range in this area is around 700 Rs.


```python
#Multi Label Binarizer is used to one hot encode the rest_type and cuisine columns as it has a list of values
#in each row and thus need to be encoded to binary values for model to use.

from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
final_rest_type = []
df_unique['rest_type'] = df_unique['rest_type'].fillna('')
for i in df_unique['rest_type']:
    if ',' in i:
        t = i.split(",")
        temp = []
        for j in t:
            j = j.strip()
            temp.append(j)
        final_rest_type.append(temp)
    else:
        temp = []
        temp.append(i)
        final_rest_type.append(temp)

arr = mlb.fit_transform(final_rest_type)
df_rest_type = (pd.DataFrame(arr, index=df_unique.index, columns=mlb.classes_))
df_rest_type.shape
```

```python
mlb1 = MultiLabelBinarizer()
final_cuisine = []
df_unique['cuisines'] = df_unique['cuisines'].fillna('')
for i in df_unique['cuisines']:
    if ',' in i:
        t = i.split(",")
        temp = []
        for j in t:
            j = j.strip()
            temp.append(j)
        final_cuisine.append(temp)
    else:
        temp = []
        temp.append(i)
        final_cuisine.append(temp)

arr1 = mlb1.fit_transform(final_cuisine)
df_cuisine = (pd.DataFrame(arr1, index=df_unique.index, columns=mlb1.classes_))
df_cuisine.shape
```

```python
df_unique = pd.concat([df_unique, df_rest_type, df_cuisine], join = 'inner', axis = 1, sort=False)
```

```python
df_unique.drop(['','rest_type','cuisines'], 1, inplace=True)
```

**Question 3**

*a Task being solved*
The task being solved is supervised regression learning.

It is supervised because we have the labelled training set. And, since the output column is continuous values, the task is regression.

*b Models to choose*
The models which can be chosen are Decision Tree Regressor, Random Forest Regressor, XGBoost.
These three models are the three moels with three different characteristics. Hence, using these models will help us in understanding the difference between the three.
Decision tree is a simple tree model. Random Forest is an ensemble model which is made up of multiple different trees. Hence it is an advanced version of decision tree. XGBoost is also an ensemble model, but it is also based on the concept of gradient boosting. Gradient boosting helps in resolving the errors made in the previous step and thus improving the model. 
Thus the choice of these three different algorithms.

*c Metrics used to evaluate the model*
The metrics which will be used to evaluate the model will be r2_score.
R^2 (coefficient of determination) score is the regression score function.

*d How do you make sure not to overfit?*
To make sure the model does not overfit, we need to use cross-validation so that the training and test samples get rotated and the model is thus trained on all different types of data. To view this, a learning curve is plotted on the training and test dataset, which will give us the visualization of how the model performs on the training set and the testing set.


```python
df_temp = pd.get_dummies(df_unique['location'], prefix = 'loc', drop_first=True)
```

```python
df_unique = pd.concat([df_unique,df_temp], axis=1)
```

```python
df_unique.fillna('')
```

```python
X = df_unique.drop(columns = ['approx_cost(for two people)','location'])
y = df_unique['approx_cost(for two people)'].copy()
```

```python
y = y.fillna(0)
```

```python
X = X.loc[:,~X.columns.duplicated()]
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score

dt_reg = DecisionTreeRegressor()
scores = cross_val_score(dt_reg, X,y,cv=5, scoring = 'r2')
print (scores.mean())
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_reg = RandomForestRegressor(n_estimators = 10)
scores = cross_val_score(rf_reg, X,y,cv=5, scoring = 'r2')
print (scores.mean())
```

```python
from sklearn.model_selection import learning_curve
train_sizes, train_scores, test_scores = learning_curve(rf_reg,X,y,cv=5,scoring='r2',n_jobs=-1,error_score = 'raise')
```

```python
train_scores_m = np.mean(train_scores, axis=1)
test_scores_m = np.mean(test_scores, axis=1)
plt.figure()
plt.plot(train_sizes, train_scores_m , color='r', label='Training')
plt.title("Learning curve to depict the performance of Train Dataset - Random Forest")
plt.legend()
plt.xlabel('Train size')
plt.ylabel('R2 score')
```

```python
plt.figure()
plt.plot(train_sizes, test_scores_m, color='b', label='Test')
plt.title("Learning curve to depict the performance of Test Dataset - Random Forest")
plt.legend()
plt.xlabel('Train size')
plt.ylabel('R2 score')
```

*e The model and its findings.*
The approach used to build the model is the cross-validation as it gives better result since it tests on different training and test datasets and then calculates the score of all and returns a mean of the score. The model gives an r2_score of 72.09% using cross-validation.
The training curve is also plotted and it shows that for all the train_sizes, the accuracy remains almost in the same range.
The model is giving an accuracy of 72% which can further be improved if the model parameters are tuned, or if a different set of features be used.
As it can be seen from the above learning curves for Random Forest, the model is overfitting as the accuracy on training set is around 95%, but it gets reduced to 70% in test set. Hence, we try XGBoost.

```python
import xgboost as xgb
from sklearn.metrics import r2_score

xgb_model = xgb.XGBRegressor()
scores = cross_val_score(xgb_model, X,y,cv=5, scoring = 'r2')
print (scores.mean())
```

```python
train_sizes, train_scores, test_scores = learning_curve(xgb.XGBRegressor(),X,y,cv=5,scoring='r2',n_jobs=-1,error_score = 'raise')
```

```python
train_scores_m = np.mean(train_scores, axis=1)
test_scores_m = np.mean(test_scores, axis=1)
plt.figure()
plt.plot(train_sizes, train_scores_m , color='r', label='Training')
plt.title("Learning curve to depict the performance of Train Dataset - XGBoost")
plt.legend()
plt.xlabel('Train size')
plt.ylabel('R2 score')
```

*g Tune the parameter*

A gridSearchCV is used to tune the hyper parameters of XGBoost, for determining better performance.
Hence, to tune the hyper parameters, a few were selected to pass in the parameter grid, from which grid search selected the best possible values for those. Below obtained are the best values. And, once we use those values in the cross-validation of XGBoost, we get an accuracy of around 74.2, which is an increase of around 2%. Hence, we can tune the parameters using GridSearchCV, and it can hence improve the model.

```python
from sklearn.model_selection import GridSearchCV

parameters = {'learning_rate': [0.05,0.1],
              'max_depth': [1,5],
              'n_estimators': [100,300,500]}

xgb_grid = GridSearchCV(estimator=xgb.XGBRegressor(), param_grid = parameters, cv=5, n_jobs =-1, scoring='r2', verbose = 3)
grid_result = xgb_grid.fit(X,y)

grid_result.best_params_
```
